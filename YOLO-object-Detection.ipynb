```python
import cv2
from ultralytics import YOLO

model = YOLO("C:\\Users\\Mark Mwema\\runs\\detect\\train2\\weights\\best.pt")
cv2.namedWindow("Testing my transfer model")
cv2.resizeWindow("Testing my transfer model",640,640)

cap = cv2.VideoCapture(0)
if not cap.isOpened():
        print("Failed to open window")
        exit()
while True:
    ret, frame = cap.read()
    if not ret:
        print("Failed to obtain the images")
    result = model(frame)
    annotated_frame = result[0].plot()
    cv2.imshow("Testing my transfer model",annotated_frame)
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

cap.release()
cv2.destroyAllWindows()
```
